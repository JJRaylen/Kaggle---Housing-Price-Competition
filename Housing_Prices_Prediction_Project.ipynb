# Sumarize - 5  Model Evaluation

```python
# creat list feature
import pandas as pd 
import numpy as np
data = pd.read_csv("./Data/train.csv")
features = ['LotArea','YearBuilt','1stFlrSF', '2ndFlrSF','FullBath', 'BedroomAbvGr','TotRmsAbvGrd']
X = data[features]
y = data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
# evaluate the model on the test set
X_train, X_valid,y_train, y_valid = train_test_split(X,y, train_size=0.8, test_size = 0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=1).fit(X_train, y_train)
# evaluate the model on the test set
print("Test set score: {:.2f}".format(dt_model.score(X_valid, y_valid)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state=1).fit(X_train, y_train)
# evaluate the model on the test set
print("Test set score: {:.2f}".format(rf_model.score(X_valid, y_valid)))
```
